## 1) Beautifoul soup

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests

### MANON ###

reviews_text = []
for page_number in range (2):
  # Replace with the URL of the reviews page you want to scrape
  reviews_page_url = f"https://www.jeuxvideo.com/jeux/ps5/jeu-1459479/avis/?p={page_number}"

  response = requests.get(reviews_page_url)

  # Check if the request was successful (ie. status_code == 200)
  if response.status_code == 200:
      # Parse the HTML "content"
      soup = BeautifulSoup(response.content, "html.parser")

      # Find all review elements that belong to p element in the source code of the website
      reviews_elements = soup.find_all("div", class_="txt-avis text-enrichi-forum")

      # Append each text of the review elements to the list review_text
      for element in reviews_elements :
        reviews_text.append(element.text)


dict_reviews = {"review":reviews_text}
df = pd.DataFrame(dict_reviews)

```

## Selenium 

```python
from selenium import webdriver
import pandas as pd
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialisation du ChromeDriver
driver = webdriver.Chrome('chromedriver/chromedriver')

# Première visite sur le site pour accepter les cookies
driver.get("https://www.jeuxvideo.com/jeux/ps5/jeu-1459479/avis/?p=0")

# Attendre que la page soit chargée
time.sleep(5)

# Localiser et cliquer sur le bouton d'acceptation des cookies
try:
    WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, '.jad_cmp_paywall_button-cookies.didomi-dismiss-button'))
    )
    cookie_button = driver.find_element(By.CSS_SELECTOR, '.jad_cmp_paywall_button-cookies.didomi-dismiss-button')
    cookie_button.click()
except Exception as e:
    print("Le bouton d'acceptation des cookies n'a pas été trouvé ou une erreur est survenue:", e)

# Attendre que la page soit actualisée après le clic
time.sleep(3)

reviews_text = []
for page_number in range(2):
    reviews_page_url = f"https://www.jeuxvideo.com/jeux/ps5/jeu-1459479/avis/?p={page_number}"
    driver.get(reviews_page_url)

    # Attendre que la page soit actualisée
    time.sleep(3)

    # Récupérer tous les avis sur la page
    reviews_elements = driver.find_elements(By.CLASS_NAME, "txt-avis")

    for element in reviews_elements:
        reviews_text.append(element.text)

driver.quit()

# Créer un DataFrame avec les avis
df = pd.DataFrame({"review": reviews_text})

```
